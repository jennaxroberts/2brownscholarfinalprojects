```python
chicken_data = open("/content/drive/My Drive/2b_CYTB_DNA.txt").read()
```

```python
alligator_data = open("/content/drive/My Drive/amnh unit 4/2c_CYTB_DNA.txt").read()
```

```python
turtle_data = open("/content/drive/My Drive/2a_CYTB_DNA.txt").read()
```

```python


dna_RNA = {"A":"U", "T": "A", "G": "C", "C":"G"}

# chicken 

RNA_chicken = "" # made to transcribe the DNA to RNA -- space to add the new bases

for item in chicken_data: #loops through each nucleotide in the histamine DNA
  RNA_chicken += dna_RNA[item] #takes whatever nucleotide it is, matches it to the key in the dictionary and adds value to the RNA

print(RNA_chicken)


# alligator 

RNA_alligator = "" # made to transcribe the DNA to RNA -- space to add the new bases

for item in alligator_data: #loops through each nucleotide in the histamine DNA
  RNA_alligator += dna_RNA[item] #takes whatever nucleotide it is, matches it to the key in the dictionary and adds value to the RNA

print(RNA_alligator)

# turtle 

RNA_turtle = ""

for item in turtle_data: #loops through each nucleotide in the histamine DNA
  RNA_turtle += dna_RNA[item] #takes whatever nucleotide it is, matches it to the key in the dictionary and adds value to the RNA

print(RNA_turtle)
```

```python
codonDict = {"UUU":"F","UUC":"F","UUA":"L","UUG":"L", "CUU":"L","CUC":"L", 
             "CUA":"L","CUG":"L","AUU":"I","AUC":"I","AUA":"I","AUG":"M",
             "GUU":"V","GUC":"V","GUA":"V","GUG":"V","UCU":"S","UCC":"S",
             "UCA":"S","UCG":"S","CCU":"P","CCC":"P","CCA":"P","CCG":"P",
             "ACU":"T","ACC":"T","ACA":"T","ACG":"T","GCU":"A","GCC":"A",
             "GCA":"A","GCG":"A","UAU":"Y","UAC":"Y","UAA":"STOP","UAG":"STOP",
             "UGA":"STOP","CAU":"H","CAC":"H","CAA":"Q","CAG":"Q","AAU":"N",
             "AAC":"N","AAA":"K","AAG":"K","GAU":"D","GAC":"D","GAA":"E",
             "GAG":"E","UGU":"C","UGC":"C","UGG":"W","CGU":"R","CGC":"R",
             "CGA":"R","CGG":"R","AGU":"S","AGC":"S","AGA":"R","AGG":"R",
             "GGU":"G","GGC":"G","GGA":"G","GGG":"G"}



#where the new values will go when translating RNA sequence to protein 

codon = ""


# chicken protein  
protein_c = "" 

for i in RNA_chicken: #loops us through the entire RNA sequence so we can translate 
  if len(codon) == 3: # for the case that the codon has 3 characters 
    protein_c += codonDict[codon] # matches the codon(key) with the amino acid(value), and appends it to the proteinA string 
    codon = "" #empty out the codon so we can move to the next block of 3 
    codon += i 
  else:
    codon += i 


# alligator 

protein_a = "" 

for i in RNA_alligator:
   if len(codon) == 3:
      protein_a += codonDict[codon] #we have accessed the dictionary using the codon key
      codon = "" #empty out the codon holder
      codon += i 
   else:
    codon += i 
    

# turtle
protein_t = ""

for i in RNA_turtle:
   if len(codon) == 3:
      protein_t += codonDict[codon] #we have accessed the dictionary using the codon key
      codon = "" #empty out the codon holder
      codon += i 
   else:
    codon += i 
    



print(protein_a)
print(protein_c)
print(protein_t)
```

```python
similarities1 = 0 
for counter, value in enumerate(protein_c):
  if protein_a[counter] == value:
    similarities1 += 1 

print(similarities1, "= number of similarities between chicken and alligator")

whole1 = min([len(protein_c), len(protein_a)])
percent1 = similarities1/whole1*100
print("percent of similarities between the chicken and alligator = ", similarities1/whole1*100, "%")


print()

similarities2 = 0 

for counter, value in enumerate(protein_t):
  if protein_a[counter] == value:
    similarities2 += 1 
print(similarities2, "= number of similarities between turtle and alligator")

whole2 = min([len(protein_t), len(protein_a)])
percent2 = similarities2/whole2*100
print("percent of similarities between the turtle and alligator = ", similarities2/whole2*100, "%")


print()

similarities3 = 0 
for counter, value in enumerate(protein_c):
  if protein_t[counter] == value:
    similarities3 += 1 
print(similarities3, "= number of similarities between turtle and chicken")


whole3 = min([len(protein_c), len(protein_t)])
percent3 = similarities3/whole1*100
print("percent of similarities between the turtle and chicken = ", similarities3/whole3*100, "%")

```
