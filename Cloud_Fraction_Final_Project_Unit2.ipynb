This code opens the data file, create a CSV reader object to read the data, and converts it into a list or lists using the built in Python list() function.



```
csvFile = open("/content/drive/My Drive/Unit2 AMNH/CloudFraction1_2002.csv", 'r') #'r' means we want to read
csvReader = csv.reader(csvFile, delimiter=',')
cf_2002 = list(csvReader)
```

```
print(cf_2002 [0])

```

```
print(len(cf_2002 [0]))
```

```
print(len(cf_2002))
```

```
width = len(cf_2002[0])

height = len(cf_2002)
```

```
grid = np.empty([height, width, 3], dtype=np.uint8) 
```

```
all_values_2002 = [] # We will fill in this empty list with all the data values as we go

for r, row in enumerate(cf_2002):
    for c, value in enumerate(row):
        if float(value) < 99999:
            all_values_2002.append(float(value)) 
        if float(value)<=0:
            grid[r,c]=[0, 43, 130]
        elif float(value)<=0.1:
            grid[r,c]=[5, 78, 227]
        elif float(value)<=0.2:
            grid[r,c]=[36, 108, 255] 
        elif float(value)<=0.3:
            grid[r,c]=[76, 126, 230]
        elif float(value)<=0.4:
            grid[r,c]=[104, 151, 247]
        elif float(value)<=0.5:
            grid[r,c]=[132, 169, 245]
        elif float(value)<=0.6:
            grid[r,c]=[166, 192, 247]
        elif float(value)<=0.7:
            grid[r,c]=[204, 220, 252]
        elif float(value)<=0.8:
            grid[r,c]=[225, 233, 250]
        elif float(value)<=0.9:
            grid[r,c]=[232, 238, 250]
        elif float(value)<=1.0:
            grid[r,c]=[247, 250, 255]
        else:
            grid[r,c]=[0,0,0]
```

```
print(min(all_values_2002))
print(max(all_values_2002))
```

```
plt.title("Cloud Fraction in 2002")
plt.imshow(grid)
plt.axis('off')
plt.savefig("Land_temperature.png",dpi = 1000)
plt.show()
```
